```python
valores = [0, 3, 4, 5]   # índice 0 dummy
pesos   = [0, 2, 3, 4]   # índice 0 dummy
n = len(valores) - 1     # número de itens (3)
C = 5                    # capacidade máxima

print(f"Número de itens: {n}")
print(f"Capacidade: {C}")
print(f"Valores: {valores[1:]}")
print(f"Pesos: {pesos[1:]}")

def knapsack_dp(valores, pesos, n, C):
    """
    Preenche a tabela de programação dinâmica para o problema da mochila.
    
    Retorna:
        tabela: matriz (n+1) x (C+1) com os valores ótimos.
    """
    # Inicializa tabela com zeros
    tabela = [[0] * (C + 1) for _ in range(n + 1)]
    
    # Preenchimento da tabela
    for i in range(1, n + 1):
        for j in range(1, C + 1):
            if pesos[i] > j:
                tabela[i][j] = tabela[i-1][j]
            else:
                tabela[i][j] = max(tabela[i-1][j], 
                                    tabela[i-1][j - pesos[i]] + valores[i])
    return tabela

# Gerar a tabela
tabela = knapsack_dp(valores, pesos, n, C)

# Exibir a tabela (opcional)
print("\nTabela de valores ótimos (linhas = itens, colunas = capacidade):")
for i in range(n+1):
    print(f"i={i}: {tabela[i]}")

def extrair_itens(tabela, pesos, n, C):
    """
    Realiza o backtracking na tabela DP para identificar quais itens
    fazem parte da solução ótima.
    
    Parâmetros:
        tabela: matriz (n+1) x (C+1) gerada pelo knapsack_dp
        pesos: lista de pesos (índice 0 dummy)
        n: número de itens
        C: capacidade da mochila
    
    Retorna:
        lista com os índices dos itens selecionados
    """
    itens = []
    capacidade_atual = C
    
    for i in range(n, 0, -1):
        # Se o valor mudou em relação à linha anterior, o item i foi incluído
        if tabela[i][capacidade_atual] != tabela[i-1][capacidade_atual]:
            itens.append(i)
            capacidade_atual -= pesos[i]
    
    itens.reverse()  # opcional: ordenar crescente
    return itens

itens_otimos = extrair_itens(tabela, pesos, n, C)
print(f"\nItens selecionados (índices): {itens_otimos}")

# %% [markdown]
# ## 4. Exibindo o resultado completo

# %%
valor_otimo = tabela[n][C]
print(f"Valor máximo obtido: {valor_otimo}")
print("Itens escolhidos:")
for idx in itens_otimos:
    print(f"  Item {idx}: valor = {valores[idx]}, peso = {pesos[idx]}")
print(f"Peso total: {sum(pesos[i] for i in itens_otimos)}")

# %% [markdown]
# ## 5. Função genérica para qualquer problema
# 
# A seguir, uma função que une os passos anteriores, retornando tanto a tabela quanto a lista de itens.

# %%
def resolver_mochila_com_itens(valores, pesos, capacidade):
    """
    Resolve o problema da mochila e retorna o valor ótimo e a lista de itens.
    
    Parâmetros:
        valores: lista de valores dos itens (índice 0 ignorado)
        pesos: lista de pesos dos itens (índice 0 ignorado)
        capacidade: inteiro C
    
    Retorna:
        (valor_otimo, itens_selecionados)
    """
    n = len(valores) - 1
    tabela = knapsack_dp(valores, pesos, n, capacidade)
    itens = extrair_itens(tabela, pesos, n, capacidade)
    return tabela[n][capacidade], itens

# Exemplo com outros dados
valores2 = [0, 10, 20, 30, 40]
pesos2   = [0, 1, 3, 4, 5]
cap2 = 8

valor_otimo2, itens2 = resolver_mochila_com_itens(valores2, pesos2, cap2)
print("\n=== Segundo exemplo ===")
print(f"Capacidade: {cap2}")
print(f"Valor ótimo: {valor_otimo2}")
print(f"Itens selecionados: {itens2}")
for i in itens2:
    print(f"  Item {i}: valor={valores2[i]}, peso={pesos2[i]}")
```
