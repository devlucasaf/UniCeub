```python
from transformers import pipeline


fillmask = pipeline('fill-mask',
                    model='bert-base-uncased'
                    # model='neuralmind/bert-large-portuguese-cased'
)
```

```python
def test_mask(text):
    results = fillmask(text)
    for res in results:
        print(f"Score: {res['score']:.4f} | Sugestão: {res['token_str']} | Frase: {res['sequence']}")
    print('-' * 20)

# Exemplos de teste
test_mask("Paris is the [MASK] of France.")
test_mask("The weather today is very [MASK].")
test_mask("Artificial Intelligence will [MASK] the world.")
```
