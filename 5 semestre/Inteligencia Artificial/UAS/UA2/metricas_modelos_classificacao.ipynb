```python
# Função para calcular métricas
def calcular_metricas(tp, fp, fn, tn):
    acuracia = (tp + tn) / (tp + tn + fp + fn)
    precisao = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = 2 * (precisao * recall) / (precisao + recall)
    
    return acuracia, precisao, recall, f1


# Dados dos classificadores
classificadores = {
    "KNN": {
        "TP": 4641, 
        "FP": 363, 
        "FN": 359, 
        "TN": 4637
    },
    "Naive Bayes": {
        "TP": 4448, 
        "FP": 582, 
        "FN": 552, 
        "TN": 4418
    },
    "SVM": {
        "TP": 4583, 
        "FP": 392, 
        "FN": 417, 
        "TN": 4608
    }
}


# Cálculo e exibição dos resultados
for nome, valores in classificadores.items():
    acuracia, precisao, recall, f1 = calcular_metricas(
        valores["TP"], valores["FP"], valores["FN"], valores["TN"]
    )
    
    print(f"\n{nome}")
    print(f"Acurácia: {acuracia:.4f}")
    print(f"Precisão: {precisao:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
```

```python
import pandas as pd
```

```python
resultados = []

for nome, valores in classificadores.items():
    acuracia, precisao, recall, f1 = calcular_metricas(
        valores["TP"], valores["FP"], valores["FN"], valores["TN"]
    )
    
    resultados.append({
        "Modelo": nome,
        "Acurácia": acuracia,
        "Precisão": precisao,
        "Recall": recall,
        "F1-Score": f1
    })

df = pd.DataFrame(resultados)
df
```
